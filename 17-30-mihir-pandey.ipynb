```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/kaggle/input/students-exam-scores/Expanded_data_with_more_features.csv")
```

```python
df.head(2)
```

```python
df.describe()
```

```python
df.info()
```

**Drop unnamed Column**

```python
df =df.drop("unnamed :0", axis =1)
```

```python
df.head()
```

**Change weekly study hours column**

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","05-10")
```

```python
df.head(2)
```

**Gender Distribution**

```python
plt.figure(figsize = (5,4.5))
ax = sns.countplot(data =df, x ="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above chart we have analysed that:
#the number of females in the data is more than the number of males.
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize =(5,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and student's Score")
plt.show()
```

```python
#from the above Treemap we can conlclude that:
#The education of the parents have a good impact on their children Score.
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize =(5,4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

```python
#From the above chart we have concluded that:
#THERE IS NO/NEGLIGIBLE IMPACT ON STUDENT'S SCORE DUE TO THEIR PARENT'S MARITAL STATUS.
```

**Distribution of Ethnic Groups**

```python
print(df["EthnicGroup"].unique())
```

```python
groupA =df.loc[(df['EthnicGroup'] == "group A")].count()
groupB =df.loc[(df['EthnicGroup'] == "group B")].count()
groupC =df.loc[(df['EthnicGroup'] == "group C")].count()
groupD =df.loc[(df['EthnicGroup'] == "group D")].count()
groupE =df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["groupA","groupB","groupC","groupD","groupE"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(mlist, labels=l, autopct='%1.2f%%')
plt.title("Distribution of Ethnic Group")
plt.show()
```
